```python
import pandas as pd
import re
```

```python
eq_data = pd.read_csv("Eartquakes-1990-2023.csv")
eq_data.head(10)
```

```python
eq_data['distance_in_km'] = eq_data['place'].str.extract(r'(\d+)').astype(float)
eq_data.head(10)
```

```python
def check_lake(place):
    if re.search(r'\bLake\b|\bLakes\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_lake'] = eq_data['place'].apply(check_lake)

```

```python
def check_volcano(place):
    if re.search(r'\bVolcano\b|\bVolcanoes\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_volcano'] = eq_data['place'].apply(check_volcano)

```

```python
def check_island(place):
    if re.search(r'\bIsland\b|\bIslands\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_island'] = eq_data['place'].apply(check_island)
```

```python
def check_river(place):
    if re.search(r'\bRiver\b|\bRivers\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_river'] = eq_data['place'].apply(check_river)
```

```python

def check_hills(place):
    if re.search(r'\bHill\b|\bHills\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_hills'] = eq_data['place'].apply(check_hills)


```

```python
def check_forest(place):
    if re.search(r'\bForest\b|\bForests\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_forest'] = eq_data['place'].apply(check_forest)
```

```python
def check_city(place):
    if re.search(r'\bCity\b|\bCities\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_city'] = eq_data['place'].apply(check_city)
```

```python
def check_ocean(place):
    if re.search(r'\bocean\b|\boceans\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_ocean'] = eq_data['place'].apply(check_ocean)
```

```python
def check_bay(place):
    if re.search(r'\bBay\b', place, re.IGNORECASE):
        return 1
    else:
        return 0

eq_data['near_bay'] = eq_data['place'].apply(check_bay)
```

```python
# Summing the number of '1's in each specified column
sum_near_lake = eq_data['near_lake'].sum()
sum_near_volcano = eq_data['near_volcano'].sum()
sum_near_mountain = eq_data['near_mountain'].sum()
sum_near_valley = eq_data['near_valley'].sum()
sum_near_island = eq_data['near_island'].sum()
sum_near_river = eq_data['near_river'].sum()
sum_near_hills = eq_data['near_hills'].sum()
sum_near_city = eq_data['near_city'].sum()
sum_near_forest = eq_data['near_forest'].sum()
sum_near_village = eq_data['near_village'].sum()
sum_near_springs = eq_data['near_springs'].sum()
sum_near_ocean = eq_data['near_ocean'].sum()
sum_near_bay = eq_data['near_bay'].sum()




# Printing the sums
print(f"Total near lake: {sum_near_lake}")
print(f"Total near volcano: {sum_near_volcano}")
print(f"Total near mountain: {sum_near_mountain}")
print(f"Total near valley: {sum_near_valley}")
print(f"Total near island: {sum_near_island}")
print(f"Total near river: {sum_near_river}")
print(f"Total near hills: {sum_near_hills}")
print(f"Total near city: {sum_near_city}")
print(f"Total near forest: {sum_near_forest}")
print(f"Total near village: {sum_near_village}")
print(f"Total near springs: {sum_near_springs}")
print(f"Total near ocean: {sum_near_ocean}")
print(f"Total near bay: {sum_near_bay}")
```

```python
missing_values_count = eq_data['distance_in_km'].isna().sum()
print("missing distances", missing_values_count)

eq_data
```

```python
#replacing NaN with o in distance_in_km
eq_data['distance_in_km'].fillna(0, inplace=True)
```

```python
# deleting place we have latitude and longitude so the directional data (SE, NW) is not useful 
eq_data.drop('place', axis=1, inplace=True)
```

```python
unique_values = eq_data['data_type'].unique()

# Print the unique values
print(unique_values)
```

```python
#counting only earthquake events 

eq1_data = eq_data.copy()

eq1_data = eq1_data[eq_data['data_type'] == 'earthquake']
eq1_data.drop('data_type', axis=1, inplace=True)

eq1_data
```

```python
eq1_data['date'] = pd.to_datetime(eq1_data['date'], format='ISO8601')
eq1_data
```

```python

# Convert state names to unique numeric codes using pandas categorization.

eq2_data = eq1_data.copy()

eq2_data['state_code'] = eq_data['state'].astype('category').cat.codes
eq2_data.drop('state', axis=1, inplace=True)
```

```python
eq2_data
```

```python
# droping unwanted features 
eq3_data = eq2_data.copy()

eq3_data.drop('time', axis=1, inplace=True)
eq3_data.drop('status', axis=1, inplace=True)
eq3_data
```

```python
column_data_types = eq3_data.dtypes

print(column_data_types)
```

```python
eq4_data = eq3_data.copy()

eq4_data['state'] = eq1_data['state']
```

```python
eq4_data
```

```python
unique_combinations = eq4_data[['state', 'state_code']].drop_duplicates().sort_values(by='state_code')

unique_combinations.to_csv('state_and_state_codes.csv', index=False)
```
