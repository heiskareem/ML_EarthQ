```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import pandas as pd

```

```python
vi_data = pd.read_csv("../preprocessed_data.csv")
```

```python
longitudes = vi_data["longitude"].tolist()
latitudes = vi_data["latitude"].tolist()

m = Basemap(projection='mill', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')

x, y = m(longitudes, latitudes)

fig = plt.figure(figsize=(12, 10))
plt.title("All affected areas")

m.plot(x, y, "o", markersize=2, color='blue')

m.drawcoastlines()
m.fillcontinents(color='coral', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawcountries()

plt.show()
```

```python
feature_data = vi_data[vi_data["near_island"] == 1]

longitudes = feature_data["longitude"].tolist()
latitudes = feature_data["latitude"].tolist()

m = Basemap(projection='mill', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')

x, y = m(longitudes, latitudes)

fig = plt.figure(figsize=(12, 10))
plt.title(f"Earthquakes Near Island")

sizes = feature_data['distance_in_km'] * 25 


m.drawcoastlines()
m.fillcontinents(color='coral', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawcountries()
m.scatter(x, y, s=sizes, color="green", marker='o', alpha=0.5)

plt.show()
```
